```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
qc_ha = QuantumCircuit(4,2)
qc_ha.x(0)
qc_ha.x(1)
qc_ha.barrier()
qc_ha.cx(0,2)
qc_ha.cx(1,2)
qc_ha.ccx(0,1,2)
qc_ha.barrier()
qc_ha.measure(2,0)
qc_ha.measure(3,1)

qc_ha.draw(output='mpl')
```

```python
counts = execute(qc_ha,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
